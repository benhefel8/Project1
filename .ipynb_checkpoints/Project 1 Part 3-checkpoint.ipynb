## Imports

```python
from composable import pipeable
from composable.strict import map, filter
import composable_records as rec
import composable_tuples as tup
from composable_glob import glob
from composable_utility import get, with_open, identity, apply
from composable_object import obj, attr
from composable_origami import fold
from operator import add
```

## Helper Functions

```python
normalize_path = lambda path: path.replace("\\", "/")
get_course = lambda path: normalize_path(path).split("/")[-2][:-2]
get_section = lambda path: normalize_path(path).split("/")[-2][-2:]
```

```python
header = [
    ("Org Defined ID", 9),
    ("UserName", 9),
    ("FirstName", 12),
    ("LastName", 12),
    ("Attempt #", 3),
    ("Attempt_Start", 20),
    ("Attempt_End", 20),
]

column_names = [col[0] for col in header]
column_widths = [col[1] for col in header]
column_names
```

## Pipe

```python
(attendance_files :=
    "attendance_example_fixed_width/*/*.txt"
    >> glob(recursive=True)
    >> map(rec.create(
             path     = lambda p: p,
             course   = lambda p: get_course(p),
             section  = lambda p: get_section(p),
             header   = lambda p: column_names,
             lines    = lambda p: with_open(lambda f: f.read().splitlines(), p),
         use_record_class=False)
          )
)>> map(rec.readable_output(num_keys=10, max_len_seq=5)) 
```
